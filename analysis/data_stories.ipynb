# Network graphs

```python
import networkx as nx
```

```python
# Helper function for plotting the degree distribution of a Graph
def plot_degree_distribution(G):
    degrees = {}
    for node in G.nodes():
        degree = G.degree(node)
        if degree not in degrees:
            degrees[degree] = 0
        degrees[degree] += 1
    sorted_degree = sorted(degrees.items())
    deg = [k for (k,v) in sorted_degree]
    cnt = [v for (k,v) in sorted_degree]
    fig, ax = plt.subplots()
    plt.bar(deg, cnt, width=0.80, color='b')
    plt.title("Degree Distribution")
    plt.ylabel("Frequency")
    plt.xlabel("Degree")
    ax.set_xticks([d+0.05 for d in deg])
    ax.set_xticklabels(deg)
```

```python
# Helper function for printing various graph properties
def describe_graph(G):
    print(nx.info(G))
    if nx.is_connected(G):
        print("Avg. Shortest Path Length: %.4f" %nx.average_shortest_path_length(G))
        print("Diameter: %.4f" %nx.diameter(G)) # Longest shortest path
    else:
        print("Graph is not connected")
        print("Diameter and Avg shortest path length are not defined!")
    print("Sparsity: %.4f" %nx.density(G))  # #edges/#edges-complete-graph
    # #closed-triplets(3*#triangles)/#all-triplets
    print("Global clustering coefficient aka Transitivity: %.4f" %nx.transitivity(G))
```

```python
# Helper function for visualizing the graph
def visualize_graph(G, with_labels=True, k=None, alpha=1.0, node_shape='o', width=1.0):
    #nx.draw_spring(G, with_labels=with_labels, alpha = alpha)
    pos = nx.spring_layout(G, k=k)
    if with_labels:
        lab = nx.draw_networkx_labels(G, pos, labels=dict([(n, n) for n in G.nodes()]))
    ec = nx.draw_networkx_edges(G, pos, alpha=alpha, width=width)
    nc = nx.draw_networkx_nodes(G, pos, nodelist=G.nodes(), node_color='r', node_shape=node_shape)
    plt.axis('off')
```

```python
df_edgelist = []
for i in range(len(df_lda_sims)):
    lda = df_lda_sims['Sims'].iloc[i]
    source = df_lda_sims['Title'].iloc[i]
    
    df_edgelist.extend([{
        'Source': source,
        'Target': df_lda_sims['Title'].iloc[j],
        'w_lda': lda[j],
    } for j in range(len(df_lda_sims)) if i != j])
```

```python
df_edgelist = pd.DataFrame(df_edgelist)
```

```python
lda = nx.from_pandas_edgelist(df_edgelist, 'Source', 'Target', edge_attr='w_lda', create_using=nx.Graph())
describe_graph(lda)
```

```python
visualize_graph(lda, with_labels=True, k=0.05, alpha=0.8, width='w_lda')
```
