```python
import json
import wikipediaapi
import pandas as pd
```

```python
wiki = wikipediaapi.Wikipedia('en')
```

```python
df = pd.read_pickle("../data/df_subset.pkl").sort_values(by='Title')
df_lda_sims = pd.read_csv("../data/df_lda_sims.csv").drop(columns="Unnamed: 0")
df_lda_topics = pd.read_csv("../data/df_lda_topics.csv").drop(columns="Unnamed: 0")
```

```python
data_panel = []
for i in range(len(df)):
    title = df['Title'].iloc[i]
    page = wiki.page(title=title)
    link = page.fullurl
    topic = df_lda_sims['Dominant Topic'].iloc[i]
    keywords = df_lda_topics.iloc[int(topic)].tolist()
    category = df['Category'].iloc[i]
    summary = page.summary
    data_panel.append({
        'title': title,
        'link': link,
        'topic': int(topic),
        'topic_keywords': keywords,
        'category': category,
        'summary': summary,
        'relations': {
            'causes': [],
            'consequences': [],
            'solutions': [],
            'problems': []
        }
    })
```

```python
with open("../data/data_panel.json", 'w+', encoding='utf-8') as f:
    json.dump(data_panel, f, ensure_ascii=False, indent=4)
```
